```python
import matplotlib.image as mpimg
from scipy.io import loadmat
def show_img_with_keypoints(img_name='060_0000'):
    x = loadmat(f'{img_name}.mat')
    points = x['pts_coord']
    print(list(zip(points[0], points[1])))
    img=mpimg.imread(f'{img_name}.png')
    imgplot = plt.imshow(img)
    plt.scatter(points[0][9], points[1][9])
    plt.show()
```
